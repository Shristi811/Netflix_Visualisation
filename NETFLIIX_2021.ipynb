```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
my_filepath = "netflix_titles.csv"
my_data = pd.read_csv(my_filepath, index_col="show_id")
```

```python
my_data.head()
```

```python
# The ratio of the Movies and TV Shows released till 2021
```

```python
fig, ax = plt.subplots(figsize=(6, 6)) 
my_data['type'].value_counts().plot(kind='pie', autopct='%1.1f%%', ax=ax,colors=['skyblue', '#f489d5'])
plt.ylabel('') 
```

```python
# Shows the analysis of the Number of releases per year
```

```python
my_data['release_year'].value_counts().sort_index().plot(kind='barh', figsize=(10, 26), title='Number of Releases by Year', color='#70b5fa')

plt.ylabel('Release Year')
plt.xlabel('Number of Titles')
plt.show()
```

```python
my_data['listed_in'].str.split(', ').explode().value_counts().head(10).plot(kind='bar', title='Top 10 Genres on Netflix', color='#f489d5')
plt.xlabel('Genre')
plt.ylabel('Number of Titles')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
my_data['country'].dropna().str.split(', ').explode().value_counts().head(10).plot(kind='barh', title='Top 10 Countries by Content Count', color='#e87e7c')
plt.xlabel('Number of Titles')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

```python
movies = my_data[my_data['type'] == 'Movie'].copy()

movies['duration_min'] = movies['duration'].str.extract('(\d+)').astype(float)

plt.figure(figsize=(10, 6))
plt.scatter(movies['release_year'], movies['duration_min'], alpha=0.4, color='#b0067f')
plt.title('Scatter Plot: Movie Duration vs Release Year')
plt.xlabel('Release Year')
plt.ylabel('Duration (minutes)')
plt.tight_layout()
plt.show()
```
